```python
# 1. 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 2. 데이터 불러오기
data = pd.read_csv(r'C:\Users\TAEKWONLEE\Downloads\trip.csv')

# 3. 결측치 처리 (예: fare_amount만 평균으로 대체)
data['fare_amount'] = data['fare_amount'].fillna(data['fare_amount'].mean())

# 4. 숫자형 컬럼 리스트 준비
num_cols = ['passenger_count', 'trip_distance', 'fare_amount', 'tip_amount', 'tolls_amount']

# 5. 이상치 확인을 위한 scatter plot 그리기
for col in num_cols:
    plt.figure(figsize=(8, 3))
    plt.scatter(data.index, data[col], alpha=0.3)
    plt.title(f'Index vs {col} (Scatter Plot)')
    plt.xlabel('Index')
    plt.ylabel(col)
    plt.show()

# 6. IQR 방법을 통한 이상치 제거 (모든 숫자형 변수에 대해 반복)
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = data.shape[0]
    data = data[(data[col] >= lower) & (data[col] <= upper)]
    after = data.shape[0]
    print(f"{col} 이상치 제거: {before - after}개 행 삭제 (남은 데이터: {after}행)")

# 7. 이상치 제거 후 데이터 구조와 통계 확인
print(data.info())
print(data.describe())
```

```python
# 1. 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 2. 데이터 불러오기
data = pd.read_csv(r'C:\Users\TAEKWONLEE\Downloads\trip.csv')

# 3. (필요 시) 결측치 처리
data['fare_amount'] = data['fare_amount'].fillna(data['fare_amount'].mean())

# 4. 탑승/하차 시간 문자열 → datetime 타입 변환
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])

# 5. 주행 시간(분) 파생 변수 생성
data['trip_duration_min'] = (data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime']).dt.total_seconds() / 60

# 6. 데이터 미리보기로 파생변수 확인
print(data[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_duration_min', 'trip_distance', 'fare_amount']].head())

# 7. 주행 시간, 거리, 요금 간 상관관계 분석
corr = data[['trip_duration_min', 'trip_distance', 'fare_amount']].corr()
print("상관계수 표:\n", corr)

# 8. 상관관계 시각화 (산점도)
plt.figure(figsize=(6, 4))
plt.scatter(data['trip_duration_min'], data['fare_amount'], alpha=0.3)
plt.xlabel('Trip Duration (min)')
plt.ylabel('Fare Amount')
plt.title('주행 시간과 요금의 관계')
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(data['trip_distance'], data['fare_amount'], alpha=0.3)
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.title('주행 거리와 요금의 관계')
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(data['trip_duration_min'], data['trip_distance'], alpha=0.3)
plt.xlabel('Trip Duration (min)')
plt.ylabel('Trip Distance')
plt.title('주행 시간과 거리의 관계')
plt.show()
```

```python
# 승객수+결제수단별 평균 요금
pivot2 = data.pivot_table(
    index='passenger_count', columns='payment_method',
    values='fare_amount',
    aggfunc='mean'
).round(2)
print(pivot2)

# 시각화
pivot2.plot(kind='bar', figsize=(10,5))
plt.ylabel('평균 요금($)')
plt.title('승객수/결제수단별 평균 요금')
plt.legend(title='결제수단')
plt.show()
```

```python
# 주행거리 구간 나누기
bins = [0, 1, 3, 6, 10, data['trip_distance'].max()]
labels = ['0-1', '1-3', '3-6', '6-10', '10+']
data['distance_bin'] = pd.cut(data['trip_distance'], bins=bins, labels=labels, include_lowest=True)

# 거리구간+결제수단별 평균 요금
pivot3 = data.pivot_table(
    index='distance_bin', columns='payment_method',
    values='fare_amount',
    aggfunc='mean'
).round(2)
print(pivot3)

# 시각화
pivot3.plot(kind='bar', figsize=(8,5))
plt.ylabel('평균 요금($)')
plt.title('주행거리 구간/결제수단별 평균 요금')
plt.legend(title='결제수단')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기 (필요시 결측치 처리, 이상치 처리 등 전처리 완료된 상태로 가정)
data = pd.read_csv(r'C:\Users\TAEKWONLEE\Downloads\trip.csv')

# 결제수단 단일화
data['payment_method'] = data['payment_method'].replace({'Credit Card': 'Card', 'Debit Card': 'Card'})

# 날짜타입 변환 및 요일 추출
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['weekday_num'] = data['tpep_pickup_datetime'].dt.dayofweek  # 0=월, 6=일
weekday_map = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
data['weekday'] = data['weekday_num'].map(weekday_map)

# 요일별+결제수단별 평균 요금 집계
pivot = data.pivot_table(
    index='weekday', columns='payment_method',
    values='fare_amount',
    aggfunc='mean'
).reindex(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])  # 요일 순서 맞추기

print("요일별+결제수단별 평균 요금")
print(pivot.round(2))

# 시각화 (요일별로 결제수단별 평균 요금 비교)
pivot.plot(kind='bar', figsize=(10,6))
plt.ylabel('평균 요금($)')
plt.title('요일별/결제수단별 평균 요금')
plt.legend(title='결제수단')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
