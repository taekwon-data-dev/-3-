```python
class Square:
    def __init__(self):
        self.square = int(input(
            '넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n1. 직사각형  2. 평행사변형  3. 사다리꼴 \n>>> '
        ))

        if self.square == 1:
            print(self.rect())

        elif self.square == 2:
            print(self.par())

        elif self.square == 3:
            print(self.trape())

        else:
            print('1, 2, 3 중에서 다시 입력해주세요.')

    def rect(self):
        width, height = map(int, input('가로, 세로를 입력하세요. 예시: 10,5\n>>> ').split(','))
        area = width * height
        return f'직사각형의 넓이는: {area}'

    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시: 8,4\n>>> ').split(','))
        area = base * height
        return f'평행사변형의 넓이는: {area}'

    def trape(self):
        top, bottom, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시: 6,10,4\n>>> ').split(','))
        area = (top + bottom) * height / 2
        return f'사다리꼴의 넓이는: {area}'
```

```python
class OrderSystem:
    def __init__(self):
        self.menu = ['아메리카노', '카페라떼', '바닐라라떼', '초코라떼', '녹차프라푸치노']
        self.price = [2000, 2500, 2800, 3000, 3500]
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0
        self.temp = ''

    def menu_select(self):
        print("===== 메뉴판 =====")
        for i, (m, p) in enumerate(zip(self.menu, self.price), start=1):
            print(f"{i}. {m} - {p}원")
        print("=================")

        n = 0
        while n < 1 or n > len(self.menu):
            try:
                n = int(input("음료 번호를 입력하세요: "))
                if 1 <= n <= len(self.menu):
                    pass
                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
            except:
                print("숫자를 입력하세요.")

        self._select_temperature()
        self.order_menu.append(self.temp + ' ' + self.menu[n - 1])
        self.order_price.append(self.price[n - 1])
        self.price_sum = self.price[n - 1]
        print(f"{self.temp} {self.menu[n - 1]} : {self.price[n - 1]}원")

        # 추가 주문 루프
        while True:
            try:
                print()
                n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
                if n == 0:
                    print("\n🧾 주문이 완료되었습니다.")
                    print("🛒 주문 목록 :", self.order_menu)
                    print("💰 가격 목록 :", self.order_price)
                    print("💵 총 합계   :", self.price_sum, "원")
                    break
                elif 1 <= n <= len(self.menu):
                    self._select_temperature()
                    self.order_menu.append(self.temp + ' ' + self.menu[n - 1])
                    self.order_price.append(self.price[n - 1])
                    self.price_sum += self.price[n - 1]
                    print(f"추가 주문: {self.temp} {self.menu[n - 1]} {self.price[n - 1]}원")
                    print(f"💰 현재 합계 : {self.price_sum}원")
                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
            except:
                print("숫자를 입력하세요.")

    def _select_temperature(self):
        t = 0
        while t != 1 and t != 2:
            try:
                t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                if t == 1:
                    self.temp = 'HOT'
                elif t == 2:
                    self.temp = 'ICE'
                else:
                    print("1과 2 중 하나를 입력하세요.")
            except:
                print("숫자를 입력하세요.")


# ✅ 실행 코드 (클래스 정의 이후에 반드시 있어야 함)
if __name__ == "__main__":
    order = OrderSystem()
    order.menu_select()
```

```python
menu_temp = ['HOT americano', 'ICE latte', 'ICE mocha', 'ICE choco_latte']
price_temp = [2000, 3000, 3000, 3000]

# 전체 너비
width = 40

# 외곽 윗부분
print('⟝' + '-' * (width - 2) + '⟞')

# 위 여백
for _ in range(2):
    print('|' + ' ' * (width - 2) + '|')

# 주문 상품 출력
for menu, price in zip(menu_temp, price_temp):
    line = f"{menu:<20} : {price:>6}원"
    padding = (width - 2 - len(line))
    print('|' + line + ' ' * padding + '|')

# 합계 금액 출력
total = sum(price_temp)
line = f"합계 금액{'':<13} : {total:>6}원"
padding = (width - 2 - len(line))
print('|' + line + ' ' * padding + '|')

# 아래 여백
for _ in range(2):
    print('|' + ' ' * (width - 2) + '|')

# 외곽 아랫부분
print('⟝' + '-' * (width - 2) + '⟞')
```
